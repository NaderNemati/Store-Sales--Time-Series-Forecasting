```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
holidays_df = pd.read_csv('/home/nnemati/forcasting_time_series/kaggle_chalenge/sale_forcasting_chalenge/data/raw/holidays_events.csv')
```

```python
holidays_df.tail(20)
```

```python
holidays_df['date'] = pd.to_datetime(holidays_df['date'], format= "%Y-%m-%d")
```

```python
holiday_counts = holidays_df['date'].value_counts()
print(f"Total number of holidays: {len(holidays_df)}")
print(holiday_counts)
```

```python
locale_type_counts = holidays_df['locale'].value_counts()
print("\nHoliday counts by locale type:")
print(locale_type_counts)
```

```python
locale_type_counts.plot(kind='bar')
plt.title('Distribution of Holidays by Locale Type')
plt.xlabel('Locale Type')
plt.ylabel('Frequency')
plt.show()
```

```python
holidays_df['month'] = holidays_df['date'].dt.month_name()
monthly_distribution = holidays_df['month'].value_counts().sort_index()
print("\nHoliday counts by month:")
print(monthly_distribution)
```

```python
monthly_distribution.plot(kind='bar')
plt.title('Monthly Distribution of Holidays')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.show()
```

```python
holiday_counts_by_region = holidays_df['locale_name'].value_counts()

holiday_counts_by_region.plot(kind='bar', figsize=(12, 6))
plt.title('Number of Holidays by Region')
plt.xlabel('Region')
plt.ylabel('Number of Holidays')
plt.show()


print(holiday_counts_by_region)
```
