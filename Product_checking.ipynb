```python
family_product_check = not missing_combinations.any()
print(f"Do all stores have all family products? {family_product_check}")
```

```python
if not family_product_check:
    missing = new_train_df[missing_combinations]
    print("Missing combinations of store_nbr and family products:")
    print(missing[['store_nbr', 'family']])
```
