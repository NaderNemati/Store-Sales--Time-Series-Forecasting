## Autocorrelation of stores up to 30 days lag

```python
# Autocorrelation
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt


for store in stores:

    tmp_aug = august_sale[august_sale['store_nbr'] == store][['date', 'agg_sales']].drop_duplicates()
    
    print('store_number: ', store)
    plt.figure(figsize=(29, 1))
    

    plot_acf(tmp_aug['agg_sales'])
    
    plt.show()
```

## Dynamic Time Warping

```python
def euclidean_distance(x, y):
    return np.abs(x - y)

for store1 in stores:
    for store2 in stores:
        if store1 != store2: 
            tmp_aug_store1 = august_sale[august_sale['store_nbr'] == store1]['agg_sales'].values
            tmp_aug_store2 = august_sale[august_sale['store_nbr'] == store2]['agg_sales'].values

            print('Store Pair: ', store1, '-', store2)
            
            distance = accelerated_dtw(tmp_aug_store1, tmp_aug_store2, dist=euclidean_distance)[0]
            
            print('DTW Distance:', distance)

```

```python
dtw_distances = {}


for store1, store2 in itertools.combinations(stores, 2):
    tmp_aug_store1 = august_sale[august_sale['store_nbr'] == store1]['agg_sales'].values
    tmp_aug_store2 = august_sale[august_sale['store_nbr'] == store2]['agg_sales'].values
    distance = accelerated_dtw(tmp_aug_store1, tmp_aug_store2, dist=euclidean_distance)[0]
    dtw_distances[(store1, store2)] = distance



sorted_pairs = sorted(dtw_distances.items(), key=lambda x: x[1])

most_similar_pairs = sorted_pairs[:10]


for pair, distance in most_similar_pairs:
    print(f"Stores {pair[0]} and {pair[1]} have a DTW distance of {distance}")
```
