average sales per stores

```python
avg_per_stores = train_df.groupby('store_nbr')['sales'].mean()
```

```python
avg_per_stores_sort = avg_per_stores.sort_values()
```

```python
plt.figure(figsize=(15,3))
print('Average sale per stores:','\n')
avg_per_stores.plot(kind='bar')
plt.xlabel('Store numbers')
plt.ylabel('Average sales')
plt.title('Average sales per store')
plt.show()
print('\n')

print('Sorted average sale per stores:','\n')
plt.figure(figsize=(15,3))
avg_per_stores_sort.plot(kind='bar')
plt.xlabel('Store numbers')
plt.ylabel('Average sales')
plt.title('Sorted average sales per store')
plt.show()
```
